##4.1 Catagorising the data

To determine of there is a relationship between the value of total nitrogen and discharge statisical tests have to be performed. To better test the data two extra columns will be added to the dataset transforming the discharge and value columns into catagorical data. The discharge will be catagorised based on the quartile range of the data.

```python
df['Discharge (ML)'].quantile([0.25, 0.5, 0.75])
```

The discharge will be split into its quartile range as ordinal data, measure as Q1 (Very low) < 7200, Q2 (Low) < 4700, Q3 (High) < 570000 < Q4 (Very High).

```python
quartiles = df['Discharge (ML)'].quantile([0.25, 0.5, 0.75])

# Create a function to assign quartile range labels
def assign_quartile_range(discharge):
  if discharge <= quartiles[0.25]:
    return 'Q1 (Very Low)'
  elif discharge <= quartiles[0.5]:
    return 'Q2 (Low)'
  elif discharge <= quartiles[0.75]:
    return 'Q3 (High)'
  else:
    return 'Q4 (Very High)'

# Apply the function to create a new column 'Discharge Quartile'
df['Discharge Quartile'] = df['Discharge (ML)'].apply(assign_quartile_range)

# Print the updated DataFrame
print(df)
```

For the purpose of statistical analysis, a column was added to indicate whether the total nitrogen value exceeds the recognized safe level of 0.75 mg/L, as recommended by the Australian Department of Climate Change, Energy, the Environment, and Water (DCCEEW). The guidelines for rivers and streams suggest a total nitrogen concentration range of 0.01 to 0.75 mg/L.(https://www.dcceew.gov.au/environment/protection/npi/substances/fact-sheets/total-nitrogen#toc_4)

```python
df['Safe levels of Total N'] = df['Value (mg per L)'].apply(lambda x: 'Safe' if x < 0.75 else 'Unsafe')
```

###4.3 Statistcal tests

###4.3.1 Chi-Square test of association

Using the catagorical columns 'Discharge Quartile' and 'Safe levels of Total N' a chi-square test will determine if there is a accociation between the Discharge and Nitrogen values.

```python
import pandas as pd
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(df['Discharge Quartile'], df['Safe levels of Total N'])

# Perform the chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Print the results
print("Chi-square statistic:", chi2)
print("P-value:", p)
print("Degrees of freedom:", dof)
print("Expected frequencies:", expected)

# Interpret the results
alpha = 0.05  # Set your significance level
if p < alpha:
  print("There is a significant association between Discharge Quartile and Safe levels of Total N.")
else:
  print("There is no significant association between Discharge Quartile and Safe levels of Total N.")
```

the p-value is less than the 0.05 significance level therefore the null hypothesis is rejected and there is a significant association between the Discharge Quartile and The Safe levels of Total N. Furter analysis will have to occur the determine the relationship between the variables.

###4.3.2 Correlation Analysis

```python
from scipy.stats import pearsonr

# Calculate the Pearson correlation coefficient and p-value
for site_name in df['Site Name'].unique():
  site_data = df[df['Site Name'] == site_name]
  if not site_data.empty:
    x = site_data['Discharge (ML)']
    y = site_data['Value (mg per L)']
    pearson_corr, pearson_p = pearsonr(x, y)
    print(f"Site Name: {site_name}")
    print("Pearson Correlation Coefficient:", pearson_corr)
    print("Pearson P-Value:", pearson_p)
    print("---")

    alpha = 0.05  # significance level
if pearson_p < alpha:
    print("Pearson: Reject the null hypothesis (significant correlation)")
else:
    print("Pearson: Fail to reject the null hypothesis (no significant correlation)")
```

The Pearson test indicated a significant relationship between the variables. A regression analysis will be conducted to determine whether this relationship is positive or negative, as well as its statistical significance.

###4.2.3 Regression Analysis

###5.5 Scatter Plot with line of best fit

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

plt.scatter(df['Discharge (ML)'], df['Value (mg per L)'])
plt.xlabel('Discharge (ML)')
plt.ylabel('Value (mg per L)')
plt.title('Scatter Plot of Discharge vs. Value')

x = df['Discharge (ML)'].values.reshape(-1, 1)
y = df['Value (mg per L)'].values
model = LinearRegression()
model.fit(x, y)
y_pred = model.predict(x)
plt.plot(x, y_pred, color='red', label='Line of Best Fit')
plt.legend()

plt.show()
```
